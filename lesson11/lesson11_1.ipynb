```python
import requests

url="https://raw.githubusercontent.com/roberthsu2003/__2025_04_29_python_2_4/refs/heads/main/lesson10/students.csv"
response=requests.get(url)
type(response)
if response.status_code==200:
    print("成功")
content:str = response.text
```

```python
print(content)
```

```python
import io
from io import StringIO
import csv
file=io.StringIO(content)
reader=csv.DictReader(file)
students=[row for row in reader]
file.close()
students
```
