```python
data=np.random.randint(50,101,size=(5,5))
df=pd.DataFrame(data,index=["學生1","學生2","學生3","學生4","學生5"],
            columns=["國文","英文","數學","地理","歷史"])
df
```

```python
#依據欄位名稱選取
df["國文"]
df[["國文","英文"]]
```

```python
sum=np.sum(df,axis=1)
mean=np.mean(df,axis=1)
df["總分"]=sum
df["平均"]=mean
```
